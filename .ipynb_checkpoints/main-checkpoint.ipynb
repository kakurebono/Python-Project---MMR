```python
from PIL import Image
from playsound import playsound
import threading
import time
from IPython.display import clear_output

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

# define rooms and items

#game room definitions

couch = {
    "name": "couch",
    "type": "furniture",
}

door_a = {
    "name": "door a",
    "type": "door",
}

key_a = {
    "name": "key for door a",
    "type": "key",
    "target": door_a,
}

piano = {
    "name": "piano",
    "type": "furniture",
}

game_room = {
    "name": "game room",
    "type": "room",
}

#bedroom 1 definitions

bedroom_1 = {
    "name": "bedroom 1",
    "type": "room",  
}

queen_bed = {
    "name": "queen bed",
    "type": "furniture",
}

door_b = {
    "name": "door b",
    "type": "door",
}

key_b = {
    "name": "key for door b",
    "type": "key",
    "target": door_b,
}

door_c = {
    "name": "door c",
    "type": "door",
}

#bedroom 2 definitions


double_bed = {
    "name": "double bed",
    "type": "furniture",
}

key_c = {
    "name": "key for door c",
    "type": "key",
    "target": door_c,
}

door_d = {
    "name": "door d",
    "type": "door",
}

key_d = {
    "name": "key for door d",
    "type": "key",
    "target": door_d,
}

dresser = {
    "name": "dresser",
    "type": "furniture",
}

bedroom_2 = {
    "name": "bedroom 2",
    "type": "room",
}

#living room definitions

living_room = {
    "name": "living room",
    "type": "room",  
}

dining_table = {
    "name": "dining table",
    "type": "furniture",
}

#outside definition

outside = {
    "name": "outside"
}

#all elements lists

all_rooms = [game_room, bedroom_1, bedroom_2, living_room, outside]

all_doors = [door_a, door_b, door_c, door_d]

# define which items/rooms are related

object_relations = {
    "game room": [couch, piano, door_a],
    "piano": [key_a],
    "outside": [door_d],
    "door a": [game_room, bedroom_1],
    "bedroom 1": [queen_bed, door_b, door_c],
    'queen bed': [key_b],
    "bedroom 2": [double_bed, door_b, dresser],
    "door b": [bedroom_2, bedroom_1],
    'double bed': [key_c],
    'dresser': [key_d],
    'door c': [bedroom_1, living_room],
    'living room': [dining_table, door_d],
    'door d': [outside, living_room]   
}

room_image = {
    'game room': 'game room.jpg',
    'bedroom 1': 'bedroom 1.jpg',
    'bedroom 2': 'bedroom 2.jpeg',
    'living room': 'living room.jpeg'
}


# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside
}
```

```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def gameover():
    
    clear_output()
    print(bcolors.BOLD + 'Your time is OVER!!!' + bcolors.ENDC)
    
def start_game():
    """
    Start the game
    """
    print(bcolors.HEADER + "You wake up on a couch and find yourself in a strange house with no windows which you have never been to before.\nYou don't remember why you are here and what had happened before.\nYou feel some unknown danger is approaching and you must get out of the house, NOW!\n" + bcolors.ENDC)
    print(bcolors.UNDERLINE + 'You have 5 minutes to finish the game.\n' + bcolors.ENDC)
    t = threading.Timer(300.0, gameover)
    t.start()
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        playsound('applause4.mp3')
        clear_output()
        print(bcolors.OKGREEN + "Congrats! You escaped the room!\n" + bcolors.ENDC)
    else:
        print(bcolors.OKBLUE + "You are now in " + room["name"]+'\n' + bcolors.ENDC)
        intended_action = input("What would you like to do? Type 'explore' or 'examine'?").strip()
        print()
        if intended_action == "explore":
            explore_room(room)
            play_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine?").strip())
        else:
            print(bcolors.FAIL + "Not sure what you mean. Type 'explore' or 'examine'.\n" + bcolors.ENDC)
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    clear_output()
    items = [i["name"] for i in object_relations[room["name"]]]
    print(bcolors.OKCYAN + "You explore the room. This is " + room["name"] + ". You find " + ", ".join(items) + '\n' + bcolors.ENDC)
    img = Image.open(room_image[room['name']])
    img.show()
    
def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            print()
            output = "You examine " + item_name + ". "
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have.\n"
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key.\n"
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    #run the puzzle for piano and dresser
                    if item['name'] == 'piano':
                        piano_puzzle()
                    if item['name'] == 'dresser':
                        dresser_puzzle()
                    #code to get the key for all items
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "You find " + item_found["name"] + ".\n"
                else:
                    output += "There isn't anything interesting about it.\n"
            print(output)
            break

    if(output is None):
        print(bcolors.FAIL + "The item you requested is not found in the current room.\n" + bcolors.ENDC)
    
    if(next_room and input("Do you want to go to the next room? Enter 'yes' or 'no': ").strip() == 'yes'):
        print()
        playsound('open_door_1.mp3')
        play_room(next_room)
    else:
        play_room(current_room)
        
#Functions for the Puzzle

def piano_puzzle ():
    playsound('fur_elise.mp3')
    answer = input('Type the Author of this song:')
    while True:
        if answer == 'beethoven' or answer == 'Beethoven':
            break
        else:
            print('Wrong answer!!! Try again!')
            piano_puzzle()
        
def dresser_puzzle () :
    '''
    this is  a slide puzzle game .
    the square marked ‘xx’ is the empty space
    the board is randomly filled so whenever you start the game you
    will get a fresh board .
    '''
    import string
    import random
    piece=()
    move = 0
    empty_space=()
    list1 = [0,1,2,3,4,5,6,7,8]
    #to get shuffled list
    random.shuffle(list1)
    print('\n'*2)
    #to make list of lists
    matrix=[]
    while list1 !=[]:
        matrix.append(list1[:3])
        list1 = list1[3:]
    # to find the empty space
    # we can also delete this section and just add a one line
    # empty_space = (x,y) before the print(‘|xx ..  ) statement
    for x in range (len(matrix)):
        for y in range(len(matrix[x])):
            if matrix[x][y] == 0:
                empty_space = (x,y)
    print()
    #to print the list in two digit matrix
    while True:
        print('\n+----+----+----|')
        for x in range (len(matrix)):
            for y in range(len(matrix[x])):
                if matrix[x][y] == 0:
                    print('|XX' , end='  ')
                else:
                     print('|' + '{:02d}' .format(matrix[x][y]), end='  ')
            print('\n+----+----+----|')
        #ask for the next move
        num = input('\nplease type the number of the piece to move : ( q ) to quit ')
        if num in ['q','Q']:
            break
        num = int(num)
        for i in range(len(matrix)):
            for j in range(len(matrix[i])):
                if num == matrix[i][j]:
                    piece = (i,j)
        if num > 8:
            print('illegal move')
        else:
            if(
empty_space==(piece[0]-1,piece[1]))\
               or(empty_space==(piece[0]+1,piece[1]))\
               or(empty_space==(piece[0],piece[1]-1))\
               or(empty_space==(piece[0],piece[1]+1)):
                matrix[empty_space[0]][empty_space[1]]=num
                matrix[piece[0]][piece[1]]=0
                empty_space=(piece[0],piece[1])
                move = move +1
                print()
                print('you have made',move , 'moves so far')
                print(2*'\n')
            else:
                print('illegal move , try again')
```
